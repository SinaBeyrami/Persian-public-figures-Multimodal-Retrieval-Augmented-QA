**Hazm & Dadmatools Preprocess**

```python
from hazm import Normalizer as HazmNormalizer
from dadmatools.normalizer import Normalizer as DadmaNormalizer
import json

# Create normalizer instances
hazm_normalizer = HazmNormalizer()
dadma_normalizer = DadmaNormalizer(
    full_cleaning=False,
    unify_chars=True,
    refine_punc_spacing=True,
    remove_extra_space=True,
    remove_puncs=False,
    remove_html=True,
    remove_stop_word=False,
    replace_email_with=None,
    replace_number_with=None,
    replace_url_with="",
    replace_mobile_number_with="",
    replace_emoji_with="",
    replace_home_number_with=""
)

# Read the JSON file
with open('/kaggle/input/final-data/corrected_data.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Function to combine normalization with both normalizers
def normalize_text_combined(text):
    if isinstance(text, str) and text.strip():
        # First, normalize with Hazm
        text = hazm_normalizer.normalize(text)
        # Then, normalize with DadmaTools
        text = dadma_normalizer.normalize(text)
        return text
    return text

# Normalize the specified fields using both methods
for item in data:
    if 'summary' in item:
        item['summary'] = normalize_text_combined(item['summary'])
    if 'summary2' in item:
        item['summary2'] = normalize_text_combined(item['summary2'])
    if 'summary3' in item:
        item['summary3'] = normalize_text_combined(item['summary3'])

# Save the normalized data to a file
with open('normalized_data.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

```
